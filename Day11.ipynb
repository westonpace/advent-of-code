```python
class Seat:
    def __init__(self, x, y):
        self.occupied_neighbors = 0
        self.occupied = False
        self.x = x
        self.y = y
        
    def __repr__(self):
        return f'S({self.x},{self.y})'
              
    @property
    def empty_seat(self):
        return not self.occupied
        
    def bump_occupied(self):
        self.occupied_neighbors = self.occupied_neighbors + 1
        
    def reset(self):
        self.occupied_neighbors = 0
        
    def tick(self):
        if not self.occupied and self.occupied_neighbors == 0:
            self.occupied = True
            return True
        elif self.occupied and self.occupied_neighbors >= 5:
            self.occupied = False
            return True
        return False
        
class Floor(object):
    
    def __init__(self, x, y):
        self.occupied = False
        self.occupied_neighbors = 0
        self.x = x
        self.y = y

    def __repr__(self):
        return f'F({self.x},{self.y})'
    
    @property
    def empty_seat(self):
        return False
    
    def bump_occupied(self):
        pass
    
    def reset(self):
        pass
    
    def tick(self):
        return False
        
seats = []

for y, line in enumerate(lines):
    for x, char in enumerate(line):
        if char == 'L':
            seats.append(Seat(x, y))
        else:
            seats.append(Floor(x, y))

def get_seat(x, y):
    return seats[y * width + x]

def get_dr_diag(x, y):
    result = []
    while x < width and y < height:
        result.append(get_seat(x, y))
        x += 1
        y += 1
    return result
        
def get_ur_diag(x, y):
    result = []
    while x < width and y >= 0:
        result.append(get_seat(x, y))
        x += 1
        y -= 1
    return result

rows = [[get_seat(x, y) for x in range(width)] for y in range(height)]
cols = [[get_seat(x, y) for y in range(height)] for x in range(width)]
diags = []
for i in range(height - 1):
    diags.append(get_dr_diag(0, i))
for i in range(width - 2):
    diags.append(get_dr_diag(i+1, 0))
for i in range(height - 1):
    diags.append(get_ur_diag(0, i+1))
for i in range(width - 2):
    diags.append(get_ur_diag(i+1, height-1))
    
all_series = rows + cols + diags
```

```python
def process_series(series, backwards=False):
    itr = reversed(series) if backwards else series
    empty = True
    for item in itr:
        if not empty:
            item.bump_occupied()
            if item.empty_seat:
                empty = True
        else:
            if item.occupied:
                empty = False

def tick():
    for seat in seats:
        seat.reset()
    for serie in all_series:
        process_series(serie, False)
        process_series(serie, True)
    changed = 0
    for seat in seats:
        if seat.tick():
            changed += 1
    return changed
```
