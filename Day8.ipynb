```python
import re

with open('day8_input.txt') as f:
    lines = [line.strip() for line in f.readlines()]
```

```python
def parse_command(cmd):
    cmd_arg = cmd.split()
    return {
        'command': cmd_arg[0],
        'argument': int(cmd_arg[1])
    }

commands = [parse_command(cmd) for cmd in lines]
commands
```

```python
cmd_pointer = 0
accumulator = 0
visited = set()
while cmd_pointer not in visited:
    cmd = commands[cmd_pointer]
    visited.add(cmd_pointer)
    cmd_pointer += 1
    if cmd['command'] == 'acc':
        accumulator += cmd['argument']
    elif cmd['command'] == 'jmp':
        cmd_pointer += cmd['argument'] - 1
print(accumulator)
```

```python
def tryrun(commands):
    cmd_pointer = 0
    accumulator = 0
    visited = set()
    action_cmd_ids_visited = []
    while cmd_pointer not in visited:
        if cmd_pointer >= len(commands):
            print(f'ACCUMULATOR AT BAIL={accumulator}')
        cmd = commands[cmd_pointer]
        visited.add(cmd_pointer)
        cmd_pointer += 1
        if cmd['command'] == 'acc':
            accumulator += cmd['argument']
        elif cmd['command'] == 'jmp':
            action_cmd_ids_visited.append(cmd_pointer - 1)
            cmd_pointer += cmd['argument'] - 1
        else:
            action_cmd_ids_visited.append(cmd_pointer - 1)
    return action_cmd_ids_visited.copy()

action_cmd_ids_visited = tryrun(commands)
print(action_cmd_ids_visited)

for index_to_flip in reversed(range(len(action_cmd_ids_visited))):
    cmd_index_to_flip = action_cmd_ids_visited[index_to_flip]
    commands_copy = commands.copy()
    if commands_copy[cmd_index_to_flip]['command'] == 'jmp':
        commands_copy[cmd_index_to_flip] = {'command': 'nop', 'argument': 0}
    else:
        commands_copy[cmd_index_to_flip] = {'command': 'jmp', 'argument': commands_copy[cmd_index_to_flip]['argument']}
    tryrun(commands_copy)
```
