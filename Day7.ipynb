```python
import re

with open('day7_input.txt') as f:
    lines = [line.strip() for line in f.readlines()]
```

```python
def sanitize(item):
    if item.endswith('.'):
        item = item[:-1]
    if item.endswith('bags'):
        item = item[:-4]
    if item.endswith('bag'):
        item = item[:-3]
    item = item.strip()
    count = int(item[:item.index(' ')])
    return count, item[item.index(' ')+1:]

rules = []
for line in lines:
    parent_children = line.split('bags contain')
    if 'no other bags' in parent_children[1]:
        children = []
        counts = []
    else:
        children = []
        counts = []
        for child in parent_children[1].split(','):
            count, name = sanitize(child)
            children.append(name)
            counts.append(count)
    rule = {
        'parent': parent_children[0].strip(),
        'children': children,
        'counts': counts
    }
    rules.append(rule)
rules
```

```python
starting_bags = []
for rule in rules:
    if 'shiny gold' in rule['children']:
        starting_bags.append(rule['parent'])
starting_bags
```

```python
noted_bags = set(starting_bags)
bags_to_consider = starting_bags
while len(bags_to_consider) > 0:
    bag = bags_to_consider.pop(0)
    noted_bags.add(bag)
    for rule in rules:
        if bag in rule['children']:
            parent = rule['parent']
            if parent not in noted_bags:
                noted_bags.add(parent)
                bags_to_consider.append(parent)
len(noted_bags)
```

```python
def get_rule_by_name(name):
    return [rule for rule in rules if rule['parent'] == name][0]

get_rule_by_name('shiny gold')
```

```python
def bfs_visit(name):
    count = 1
    rule = get_rule_by_name(name)
    children = rule['children']
    if not children:
        print(f'{name} is 1 bag')
        return count
    counts = rule['counts']
    for i in range(len(children)):
        total_for_child = counts[i] * bfs_visit(children[i])
        print(f'{counts[i]} {children[i]} bags means {total_for_child}')
        count += total_for_child
    print(f'{name} has {count} bags')
    return count

bfs_visit('shiny gold')
```
