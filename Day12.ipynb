```python
import math
import re

with open('day12_input.txt') as f:
    lines = [line.strip() for line in f.readlines()]
```

```python
commands = []
for line in lines:
    match = re.match('^(\D+)(\d+)$', line)
    commands.append({
        'command': match.group(1),
        'amount': int(match.group(2))
    })
commands
```

```python
x_pos = 0
y_pos = 0
direction = 0

def patch_direction():
    global direction
    while direction < 0:
        direction += 360
    while direction >= 360:
        direction -= 360

def move(cmd, amt):
    global x_pos, y_pos, direction
    print(f'Start {x_pos},{y_pos}@{direction}', end='')
    print(f' cmd={cmd} amt={amt} End ', end='')
    if cmd == 'N':
        y_pos += amt
    elif cmd == 'E':
        x_pos += amt
    elif cmd == 'S':
        y_pos -= amt
    elif cmd == 'W':
        x_pos -= amt
    elif cmd == 'F':
        if direction == 0:
            x_pos += amt
        elif direction == 90:
            y_pos -= amt
        elif direction == 180:
            x_pos -= amt
        else:
            y_pos += amt
    elif cmd == 'L':
        direction -= amt
    elif cmd == 'R':
        direction += amt
    else:
        raise Exception('Unrecognized cmd ' + cmd)
    patch_direction()
    print(f'{x_pos},{y_pos}@{direction}')
    
for command in commands:
    move(command['command'], command['amount'])
    
print(f'x_pos={x_pos} y_pos={y_pos} direction={direction}')
```

```python
waypoint_x = 10
waypoint_y = 1
x_pos = 0
y_pos = 0

def fix_direction(amt):
    while amt < 0:
        amt += 360
    while amt >= 360:
        amt -= 360
    return amt

def move(cmd, amt):
    global x_pos, y_pos, waypoint_x, waypoint_y
    print(f'Start {x_pos},{y_pos}@{waypoint_x},{waypoint_y}', end='')
    print(f' cmd={cmd} amt={amt} End ', end='')
    if cmd == 'N':
        waypoint_y += amt
    elif cmd == 'E':
        waypoint_x += amt
    elif cmd == 'S':
        waypoint_y -= amt
    elif cmd == 'W':
        waypoint_x -= amt
    elif cmd == 'F':
        x_distance = waypoint_x * amt
        y_distance = waypoint_y * amt
        x_pos += x_distance
        y_pos += y_distance
    elif cmd == 'L':
        amt = fix_direction(amt)
        if amt == 90:
            if waypoint_x > 0 and waypoint_y > 0:
                swap = waypoint_x
                waypoint_x = -1 * waypoint_y
                waypoint_y = swap
            elif waypoint_x > 0 and waypoint_y < 0:
                swap = waypoint_x
                waypoint_x = -1 * waypoint_y
                waypoint_y = swap
            elif waypoint_x < 0 and waypoint_y < 0:
                swap = waypoint_x
                waypoint_x = -1 * waypoint_y
                waypoint_y = swap
            else:
                swap = waypoint_x
                waypoint_x = -1 * waypoint_y
                waypoint_y = swap
        elif amt == 180:
            waypoint_x = -1 * waypoint_x
            waypoint_y = -1 * waypoint_y
        elif amt == 270:
            if waypoint_x > 0 and waypoint_y > 0:
                swap = waypoint_x
                waypoint_x = waypoint_y
                waypoint_y = -1 * swap
            elif waypoint_x > 0 and waypoint_y < 0:
                swap = waypoint_x
                waypoint_x = waypoint_y
                waypoint_y = -1 * swap
            elif waypoint_x < 0 and waypoint_y < 0:
                swap = waypoint_x
                waypoint_x = waypoint_y
                waypoint_y = -1 * swap
            else:
                swap = waypoint_x
                waypoint_x = waypoint_y
                waypoint_y = -1 * swap
    elif cmd == 'R':
        amt = fix_direction(amt)
        if amt == 90:
            if waypoint_x > 0 and waypoint_y > 0:
                swap = waypoint_x
                waypoint_x = waypoint_y
                waypoint_y = -1 * swap
            elif waypoint_x > 0 and waypoint_y < 0:
                swap = waypoint_x
                waypoint_x = waypoint_y
                waypoint_y = -1 * swap
            elif waypoint_x < 0 and waypoint_y < 0:
                swap = waypoint_x
                waypoint_x = waypoint_y
                waypoint_y = -1 * swap
            else:
                swap = waypoint_x
                waypoint_x = waypoint_y
                waypoint_y = -1 * swap
        elif amt == 180:
            waypoint_x = -1 * waypoint_x
            waypoint_y = -1 * waypoint_y
        elif amt == 270:
            if waypoint_x > 0 and waypoint_y > 0:
                swap = waypoint_x
                waypoint_x = -1 * waypoint_y
                waypoint_y = swap
            elif waypoint_x > 0 and waypoint_y < 0:
                swap = waypoint_x
                waypoint_x = -1 * waypoint_y
                waypoint_y = swap
            elif waypoint_x < 0 and waypoint_y < 0:
                swap = waypoint_x
                waypoint_x = -1 * waypoint_y
                waypoint_y = swap
            else:
                swap = waypoint_x
                waypoint_x = -1 * waypoint_y
                waypoint_y = swap
    else:
        raise Exception('Unrecognized cmd ' + cmd)
    print(f'{x_pos},{y_pos}@{waypoint_x},{waypoint_y}')
    
for command in commands:
    move(command['command'], command['amount'])
    
print(f'x_pos={x_pos} y_pos={y_pos} direction={direction}')
```
