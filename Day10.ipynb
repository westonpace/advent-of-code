```python
with open('day10_input.txt') as f:
    lines = [line.strip() for line in f.readlines()]
```

```python
adapters = [int(line) for line in lines]
adapters = sorted(adapters)
adapters
```

```python
all_adapters = [0] + adapters + [adapters[-1]+3]
differences = []
for i in range(len(all_adapters)-1):
    differences.append(all_adapters[i+1] - all_adapters[i])
num_ones = len([x for x in differences if x == 1])
num_threes = len([x for x in differences if x == 3])
print(f'There are {num_ones} 1s and {num_threes} 3s')
```

```python
num_ones * num_threes
```

```python
possibilities = [0] * len(all_adapters)
possibilities[-1] = 1
print(len(all_adapters))
for i in reversed(range(len(all_adapters))):
    j = 1
    possibilities_from_here = possibilities[i]
    while i+j < len(all_adapters) and (all_adapters[i+j] - all_adapters[i]) < 4:
        print(f'i={i} j={j} i+j={i+j} possibilities[92]={possibilities[92]} possibilities[i+j]={possibilities[i+j]}')
        possibilities_from_here += possibilities[i+j]
        j += 1
    possibilities[i] = possibilities_from_here
possibilities
```
