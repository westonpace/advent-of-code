```python
import re

with open('day2_input.txt') as f:
    password_records = f.readlines()
```

```python
policies = []
passwords = []
for record in password_records:
    m = re.search('^(\d+)\-(\d+) (.): (.*)$', record.strip())
    policies.append({
        'min': int(m.group(1)),
        'max': int(m.group(2)),
        'letter': m.group(3)
    })
    passwords.append(m.group(4))    
```

```python
valid = 0

for index in range(len(policies)):
    password = passwords[index]
    policy = policies[index]
    num_occurrences = password.count(policy['letter'])
    if num_occurrences >= policy['min'] and num_occurrences <= policy['max']:
        valid = valid + 1
```

```python
print(valid)
```

```python
valid = 0

for index in range(len(policies)):
    password = passwords[index]
    policy = policies[index]
    first_matches = password[policy['min'] - 1] == policy['letter']
    second_matches = password[policy['max'] - 1] == policy['letter']
    if (first_matches and not second_matches) or (not first_matches and second_matches):
        valid = valid + 1
        print(f'{policy["min"]} {policy["max"]} {policy["letter"]} {password}')
print(valid)
```
