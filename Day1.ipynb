```python
import bisect

with open('day1_input.txt') as f:
    expense_report = f.readlines()
```

```python
expense_report
```

```python
expenses = [int(expense.strip()) for expense in expense_report]
expenses
```

```python
sorted_expenses = sorted(expenses)
sorted_expenses
```

```python
for value in sorted_expenses:
    inverse_value = 2020 - value
    possible_index = bisect.bisect_left(sorted_expenses, inverse_value)
    if possible_index < len(sorted_expenses) and sorted_expenses[possible_index] == inverse_value:
        print(value * inverse_value)
        break
```

```python
def solve_two(sorted_values, desired):
    for value in sorted_values:
        inverse_value = desired - value
        possible_index = bisect.bisect_left(sorted_values, inverse_value)
        if possible_index < len(sorted_values) and sorted_values[possible_index] == inverse_value:
            return value,inverse_value
    return None

solve_two(sorted_expenses, 2020)
```

```python
for value in sorted_expenses:
    remainder = 2020 - value
    solution = solve_two(sorted_expenses, remainder)
    if solution:
        print(f'{value},{solution[0]},{solution[1]}')
        print(value * solution[0] * solution[1])
        break
```
