```python
import re

with open('day5_input.txt') as f:
    boarding_passes = [line.strip() for line in f.readlines()]
```

```python
row_ids = []
for boarding_pass in boarding_passes:
    numeric_version = int(boarding_pass.replace('F', '0').replace('B', '1').replace('R', '1').replace('L', '0'), 2)
    row_ids.append(numeric_version)
max(row_ids)
```

```python
sorted_row_ids = sorted(row_ids)
seat_num = sorted_row_ids[0] + 1
last = sorted_row_ids[-1]
index = 1
while index < len(row_ids) - 1:
    print(f'index={index} seat_num={seat_num} sorted_row_ids[index]={sorted_row_ids[index]}')
    if sorted_row_ids[index] == seat_num:
        seat_num += 1
        index += 1
    else:
        if sorted_row_ids[index - 1] == seat_num - 1 and sorted_row_ids[index + 1] == seat_num + 1:
            print(seat_num)
        else:
            print('CANDIDATE')
        seat_num += 1
        
```
